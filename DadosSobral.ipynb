```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Adicionamos nossos dados a uma pagina do github e a importamos para desenvolvimento do codigo
df = pd.read_excel("https://raw.githubusercontent.com/Anabia-05/DadosEstatistica/main/Base_dados_sobral%20.xlsx")
```

```python
df.head(10)
```

```python
dependencia_significado ={
    1 : 'Federal',
    2 : 'Estadual',
    3: 'Municipal',
    4: 'Privada'
}

df['TP_DEPENDENCIA'] = df['TP_DEPENDENCIA'].replace(dependencia_significado)
```

```python
df.head()
```

```python
frequencia = df['TP_DEPENDENCIA'].value_counts().sort_values(ascending=True)
porcentagem = df['TP_DEPENDENCIA'].value_counts(normalize = True)* 100
Freq_TPDEPEND = pd.DataFrame({'Freq' : frequencia , '%' : porcentagem})
```

```python
Freq_TPDEPEND
```

```python
Freq_TPDEPEND.plot(kind='bar', y='Freq', legend=False)

plt.xlabel('TP_DEPENDENCIA')
plt.ylabel('Frequência')

plt.show()
```

Ao analisar os dados, observa-se uma predominância significativa das escolas municipais na amostra. São 84 escolas municipais (≈64,62%) e 27 escolas privadas (≈20,77%), totalizando cerca de 85,39% das instituições analisadas. Das escolas restantes, 18 são estaduais (≈13,85%) e apenas 1 é federal (≈0,77%). Esses dados evidenciam a predominância das escolas municipais na amostra, seguidas pelas privadas, com uma menor participação das estaduais e quase mínima das federais.


```python
localizao_dicionario = {
    1: 'Urbana',
    2: 'Rural'
}
df['TP_LOCALIZACAO'] = df['TP_LOCALIZACAO'].replace(localizao_dicionario)
```

```python
frequencia = df['TP_LOCALIZACAO'].value_counts().sort_values(ascending=True)
porcentagem = df['TP_LOCALIZACAO'].value_counts(normalize = True)* 100
Freq_TPLoc = pd.DataFrame({'Freq' : frequencia , '%' : porcentagem})
```

```python
Freq_TPLoc
```

```python
Freq_TPLoc.plot(kind='bar', y='Freq', legend=False)

plt.xlabel('TP_LOCALIZACAO')
plt.ylabel('Frequência')

plt.show()
```

Analisando os dados de localização das escolas de Sobral, observa-se uma predominância de instituições em áreas urbanas. Das 130 escolas analisadas, 120 (cerca de 92,31%) estão em áreas urbanas, enquanto apenas 10 (aproximadamente 7,69%) estão em áreas rurais. Esses números ressaltam uma concentração significativa de escolas em regiões urbanas em comparação com as rurais.
